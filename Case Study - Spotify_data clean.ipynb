```python
#  Handle Missing Values 
data = data.dropna(subset=["artist_name"])
data = data.dropna(subset=["artist_genres"])
data.isnull().sum()
```

```python
# Any duplicates in data
duplicates = data.duplicated()
print(duplicates.sum())
```

```python
#  Outlier Detection using IQR (track_duration_min)
Q1 = data["track_duration_min"].quantile(0.25)
Q3 = data["track_duration_min"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data["track_duration_min"] < lower_bound) | (data["track_duration_min"] > upper_bound)]

print("Number of Outliers:", outliers.shape[0])
```

```python
# Remove Outliers
data_clean = data[(data["track_duration_min"] >= lower_bound) & (data["track_duration_min"] <= upper_bound)]

print("Original Shape:", data.shape)
print("New Shape:", data_clean.shape)
```

```python
#  Boxplot After Removing Outliers
plt.figure(figsize=(6,4))
plt.boxplot(data_clean["track_duration_min"])
plt.title("Boxplot - Track Duration (Cleaned)")
plt.show()
```

```python
## Skewness Calculation
numeric_cols = data_clean.select_dtypes(include=["int64", "float64"]).columns
skew_values = data_clean[numeric_cols].skew().sort_values(ascending=False)

print("Skewness of numeric columns:\n")
print(skew_values)
```

```python
# Skewness of Numeric Columns
plt.figure(figsize=(6,4))
sns.barplot(x=skew_values.index, y=skew_values.values)
plt.xticks(rotation=90)
plt.xlabel("Columns")
plt.ylabel("Skewness")
plt.title("Skewness of Numeric Columns")
plt.show()
```

# Genre Analysis

```python
# Split genres
genres = data['artist_genres'].dropna().str.split(', ')
all_genres = genres.explode()

top_genres = all_genres.value_counts().head(10)
top_genres
```

```python
# Top 10 Genres
plt.figure(figsize=(10,5))
top_genres.plot(kind='bar')
plt.title("Top 10 Genres")
plt.show()
```

```python
# Top Genres by count (bar chart)
data['track_popularity'].plot(kind='hist', bins=10, edgecolor ="w")
plt.title("Track Popularity Distribution")
plt.xlabel("Popularity")
plt.ylabel("Count")
plt.show()
```

```python
# Artist Popularity vs Track Popularity
sns.scatterplot(
    data=data,
    x='artist_popularity',
    y='track_popularity',
    size='artist_followers',
    hue='explicit'
)

plt.title("Artist Popularity vs Track Popularity")
plt.show()
```

```python
# Artist Followers
top_artists = data.groupby('artist_name')['artist_followers'].max().sort_values(ascending=False)
top_10_artists = top_artists.head(10)

plt.figure(figsize=(8, 5)) 
top_10_artists.plot(kind='bar')
plt.title("Artist Followers")
plt.ylabel("Followers")
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```
