# Iterate over each page and scrape the data

```python
## Create lists for to store the data
title_data = []
rating_data = []
review_data = []
feature_data = []
original_price_data = []
discount_percentage_data =[]
discount_price_data = []

## Iterate through each page
for i in range(1,45):
    # Mention the URL
    url = f"https://www.flipkart.com/search?q=AC&page={i}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
    }

    # Fetch the HTML content from the link
    response = requests.get(url, headers = headers)

    # Break the loop in case there is a 403 error
    if response.status_code  == 403:
        print("That's a 403 Error!")
        break
        
    # Parse the HTML content
    html_code = response.text
    soup = BeautifulSoup(html_code)
    
    # Use find_all function to extract the data and append data to list
    # Product title
    title = soup.find_all("div", attrs = {"class" : "RG5Slk"})
    for t in title:
        title_data.append(t.text)
    
    # product rating
    ratings = soup.find_all("div", attrs = {"class" : "MKiFS6"})
    for r in ratings:
        rating_data.append(r.text)
    
    # product reviews
    reviews = soup.find_all("span", attrs = {"class" : "PvbNMB"})
    for r in reviews:
        review_data.append(r.text)
    
    # product feature list
    features = soup.find_all("ul", attrs = {"class" : "HwRTzP"})
    for f in features:
        feature_data.append(f.text)

    # Product Original Price
    original_price = soup.find_all("div" , attrs = {"class" : "kRYCnD gxR4EY"})
    for o in original_price:
        original_price_data.append(o.text)
        
    # Product discount Percentage
    discount_percentage = soup.find_all("div" , attrs = {"class" : "HQe8jr"})
    for d in discount_percentage:
        discount_percentage_data.append(d.text)
        
    # Product discount price
    discount_price = soup.find_all("div" , attrs = {"class" : "hZ3P6w DeU9vF"})
    for d in discount_price:
        discount_price_data.append(d.text)
```

```python
# Length of Data
len(title_data),len(rating_data),len(review_data),len(feature_data),len(original_price_data),len(discount_percentage_data),len(discount_price_data)
```

## Iterate over each page and scrape the data again(with correct way)

```python
import numpy as np

## Create lists for all features to store the values
title_data = []
rating_data = []
review_data = []
feature_data = []
original_price_data = []
discount_percentage_data =[]
discount_price_data = []

## Iterate through each page
for i in range(1,45):
    # Mention the URL
    url = f"https://www.flipkart.com/search?q=AC&page={i}"
   
    #print(url)
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
    }

    #Fetch the HTML content from the link
    response = requests.get(url, headers = headers)

     
    soup = BeautifulSoup(response.text)
    
    # Use Containerization
    containers = soup.find_all("a", attrs = {"class":"k7wcnx"})
    for c in containers:
        
        # Product title
        title = c.find("div", attrs = {"class" : "RG5Slk"})
        if title is None:
            title_data.append(np.nan)
        else:
            title_data.append(title.text)
        
        # product rating
        ratings = c.find("div", attrs = {"class" : "MKiFS6"})
        if rating is None:
            rating_data.append(np.nan)
        else:
            rating_data.append(rating.text)

        # product reviews
        review = c.find("span", attrs = {"class" : "PvbNMB"})
        if review is None:
            review_data.append(np.nan)
        else:
            review_data.append(review.text)
        
        # product feature list
        features = c.find("ul", attrs = {"class" : "HwRTzP"})
        if feature is None:
            feature_data.append(np.nan)
        else:
            feature_data.append(feature.text)
            
        # Product Original Price
        original_price = c.find("div" , attrs = {"class" : "kRYCnD gxR4EY"})
        if original_price is None:
            original_price_data.append(np.nan)
        else:
            original_price_data.append(original_price.text)
            
        # Product discount Percentage
        discount_percentage = c.find("div" , attrs = {"class" : "HQe8jr"})
        if discount_percentage is None:
            discount_percentage_data.append(np.nan)
        else:
            discount_percentage_data.append(discount_percentage.text)
            
        # Product discount price
        discount_price = c.find("div" , attrs = {"class" : "hZ3P6w DeU9vF"})
        if discount_price is None:
            discount_price_data.append(np.nan)
        else:
            discount_price_data.append(discount_price.text)
```
